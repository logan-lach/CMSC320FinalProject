```python
colors = ['beige', 'lightblue', 'black', 'lightgrey', 'purple', 'darkblue', 'purple', 'beige', 'gray', 'darkpurple', 'pink', 'black', 'darkgreen', 'orange', 'green']
i = 0
mapping = {}

for item in set(df['SURF_COND_DESC']):
    mapping[item] = colors[i]
    i += 1

    
```

```python
i = 0

items = np.where((df['LANE_DESC'] == 'Acceleration Lane') | (df['LANE_DESC'] == 'Crossover Area') | (df['LANE_DESC'] == 'Deceleration Lane'))
len(df.loc[items])

for k, v in df.loc[items].iterrows():
    item = mapping[v["SURF_COND_DESC"]]
    folium.Marker(location=[v["LATITUDE"], v["LONGITUDE"]],
                    icon=folium.Icon(color=item)).add_to(chart)

chart
    
```

```python
carroll_county_coords = [39.5423, -77]
chart = folium.Map(location=carroll_county_coords, zoom_start = 10.5)

# Ensure you're handing it floats
df['LATITUDE'] = df['LATITUDE'].astype(float)
df['LONGITUDE'] = df['LONGITUDE'].astype(float)

# Filter the DF for rows, then columns, then remove NaNs
heat_df = df[df['YEAR']==2021] # Reducing data size so it runs faster
heat_df = heat_df[['LATITUDE', 'LONGITUDE']]
heat_df = heat_df.dropna(axis=0, subset=['LATITUDE','LONGITUDE'])

# List comprehension to make out list of lists
heat_data = [[row['LATITUDE'],row['LONGITUDE']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(chart)

# Display the map
chart
```
